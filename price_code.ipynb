```python
import pandas as pd
```

```python
data = pd.DataFrame(
    {'Demand': [40, 50, 46, 42, 60, 64, 60, 56, 34, 49],
     'Price': [500, 350, 400, 450, 230, 200, 250,310,390, 420 ]

    }
)
```

```python
data
```

```python
data["% Change in Demand"] = data['Demand'].pct_change()
data["% Change in Price"] = data['Price'].pct_change()
```

```python
data
```

```python
data['Price Elasticity'] = data["% Change in Demand"] / data["% Change in Price"]
```
